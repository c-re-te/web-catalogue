```python
import pandas as pd

df = pd.read_csv('../../assets/data/sample-query.csv')
print(df.columns)
```

```python
df_leaflet = df[['id', 'autore-0-name', 'autore-0-rif', 'soggetto', 'loc-0-comune', 'loc-0-lat', 'loc-0-long']]
print(df_leaflet)
```

```python
from pprint import pprint
import numpy as np

loc_dict = {}

for idx, row in df_leaflet.iterrows():
    if not np.isnan(row['loc-0-lat']) and not np.isnan(row['loc-0-long']):
        if row['loc-0-comune'] not in loc_dict:
            loc_dict[row['loc-0-comune']] = {
                'lat': row['loc-0-lat'],
                'long': row['loc-0-long'],
                'opere_id': [row['id']]
            }
        else:
            loc_dict[row['loc-0-comune']]['opere_id'].append(row['id'])

pprint(loc_dict)
```

```python
for city in loc_dict:
    loc_dict[city]['count'] = len(loc_dict[city]['opere_id'])

pprint(loc_dict)
```
